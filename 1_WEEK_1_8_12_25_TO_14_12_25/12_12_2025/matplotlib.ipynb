### Line chart

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("flight.csv")
df.head()
```

```python
plt.plot(df['dep_time'])
plt.show()
```

```python
plt.bar(df['dep_time'],df['sched_dep_time'])
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Title")
plt.show()
```

```python
plt.hist(df['dep_time'])
plt.show()
```

```python
plt.scatter(df['dep_time'],df['sched_dep_time'])
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Title")
plt.show()
```

```python

label  = ["10","20","30","40"]
size = [1,2,3,4]
plt.pie(size,labels=label)
plt.show()
```
