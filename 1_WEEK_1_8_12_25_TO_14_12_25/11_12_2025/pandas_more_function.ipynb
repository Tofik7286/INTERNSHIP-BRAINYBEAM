### counting null values

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Movie-Data.csv")
df.isnull()
## true means that field is null
df.isnull().sum()
```

```python
df['Metascore'].isnull().sum()
```

### dropping column

```python
df = df.drop(['Rating'],axis=1)
```

```python
df
```

##### 2nd way to delete column. we have inplace attribute. set it to True it will delete column permanently.

```python
df.drop(['Rank'],axis=1,inplace=True)
```

```python
df
```

##### drop all null values rows

```python
df.dropna()
```

```python
df
```
