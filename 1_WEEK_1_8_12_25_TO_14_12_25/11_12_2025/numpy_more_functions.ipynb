```python
import numpy as np
```

### indexing and slicing

```python
arr=  np.arange(100,200,5).reshape((4,5))
arr
```

```python
print(arr[1,3])
```

```python
print(arr[1:3,1:4])
```

### mathematical operations

```python
arr1 = np.arange(20).reshape(4,5)
arr1
```

```python
arr2 = np.arange(100,200,5).reshape(4,5)
arr2
```

```python
print(arr1+arr2)
```

```python
print(arr1-arr2)
```

```python
print(arr1*arr2)
```

```python
print(arr1/arr2)
```

```python
print(arr2+100)
```

### statistical function (Mean, Median, Mode, etc...)

```python
print(np.mean(arr1))
print(np.median(arr1))
print(np.min(arr1))
print(np.max(arr1))


```

```python
print(np.cos(arr1))
```
