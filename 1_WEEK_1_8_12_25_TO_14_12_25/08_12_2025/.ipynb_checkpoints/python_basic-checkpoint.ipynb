```python
# Cell 3: Working with different data types (integer, float, and complex)
integer_variable = 10
float_variable = 365.45
complex_variable = 10 + 2j

# Display values and types
print(f"""
    integer value : {integer_variable}, type : {type(integer_variable)}
    float value : {float_variable}, type : {type(float_variable)}
    complex value : {complex_variable}, type : {type(complex_variable)}
""")
```

```python
# Cell 4: String operations
str = "Welcome to Python"

# Print whole string
print(str)

# Print only first character of string
print(str[0])

# Print 3rd character to -1 character of string using slicing
print(str[3:-1])

# Print string from 4th character to the end of string using slicing
print(str[4:])

# Print whole string 5 times
print(str * 5)
```

```python
# Cell 5: List operations
festivals = ['Navratri', 'Diwali', 'Holi', 'Rakshabandhan', 'Bakri Eid', 'Muharram']

# Print whole list
print(festivals)

# Print only first element of list
print(festivals[0])

# Print elements starting from 2nd till 3rd
print(festivals[2:4])

# Print elements starting from 2nd element till last
print(festivals[2:])

# Print whole list 4 times
print(festivals * 4)
```

```python
# Cell 6: Dictionary operations
fruit = {
    "Apple": "Apple is a healthy fruit",
    "Orange": "Orange has rich vitamin C",
}

# Display keys and values separately
for key in fruit:
    print(f"{key}: {fruit[key]}")
```

```python
# Cell 7: Tuple operations
vegetables = ('Potato', 'Brinjal', 'Tomato', 'Cabbage', 'Cauliflower')

# Print whole tuple
print(vegetables)

# Print only first element of tuple
print(vegetables[0])

# Print elements starting from 2nd till 4th
print(vegetables[2:4])

# Print elements starting from 2nd element till last
print(vegetables[2:])

# Print whole tuple twice
print(vegetables * 2)
```

```python
# 8. Write a Python program which will have Main Menu for selecting Elective Subjects as
# follows:
# Main Menu:
# 1. Joomla
# 2. Ruby onRails
# 3. Drupal
# 4. Android
# 5. iOS
# Display proper message for every choice. Use elif to create a menu:

choice = int(input("""
    Enter 1 : Joomla
    Enter 2 : Ruby onRails
    Enter 3 : Drupal
    Enter 4 : Android
    Enter 5 : IOS
      """))
if(choice == "1"):
    print("You selected Joomla")
elif choice == 2:
    print("You selected Ruby")
elif choice == 3:
    print("You selected Drupal")
elif choice == 4:
    print("You selected Android")
elif choice == 5:
    print("You selected IOS")
else : 
    print("Please choose correct option")
```
