### zeros function
 <b> this function create array of zeros</b>

```python
zero_func = np.zeros(10,dtype=int)
zero_func

# we can also convert data type of zeros function

zero_func_float = zero_func.astype(float)
zero_func_float

## we can also create 2d arrays

zero_func_2d = np.zeros((4,4))
zero_func_2d
```

### ones function

**this function create arrays of 1**

```python
ones_func = np.ones(10,dtype=int)
ones_func

# we can also convert data type of ones function

ones_func_float = ones_func.astype(float)
ones_func_float

## we can also create 2d arrays

ones_func_2d = np.ones((4,4))
ones_func_2d
```

##### linspace() function in NumPy returns an array of evenly spaced numbers over a specified range. Unlike the range() function in Python that generates numbers with a specific step size. linspace() allows you to specify the total number of points you want in the array, and NumPy will calculate the spacing between the numbers automatically.



```python
np.linspace(1,100,50,dtype=int)
```

### 2d array slicing

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
arr
```

```python
print(arr[0,1])
print(arr[1,1])
print(arr[:,1])
print(arr[1,:])
```

### mathematical functions

```python
arr = np.arange(10,50,5)
```

```python
print(arr)
print(sum(arr))
print(np.mean(arr))
print(np.median(arr))
print(np.std(arr))
print(np.var(arr))
print(arr.itemsize)
```
