## some in built functions

```python
zero_function = np.zeros(10)
zero_function


## if we want to perform 2d array we hae to use this.add()

zero_function_2d_matrix = np.zeros((4,5))
# 4 -> no. of rows
# 5 -> no. of columns

print(zero_function_2d_matrix)
print(zero_function_2d_matrix.ndim)
print(zero_function_2d_matrix.shape)
```

#### to check data type of array we have dtype function

```python
print(zero_function_2d_matrix.dtype)
print(arr3.dtype)
```

#### convert data type we use astype(data type name)

```python
new_arr = zero_function_2d_matrix.astype(int)
print(new_arr)
print(new_arr.dtype)
print(new_arr.shape)
```

### Identity function. 
##### This function will create diagonal matrix. 

```python
identity_matrix = np.identity(10)
identity_matrix_int = identity_matrix.astype(int)
print(identity_matrix_int)
```

### copy array

```python
# 1st way is normal python

arr = np.ones((4,5),dtype=int)
copy_array = arr
copy_array

#2nd way using numpy

original_array = np.zeros((5,5))
copied_array = np.copy(original_array)
print(copied_array)
```
